# The datamodel for the netflow graph

```python
from builtins import object
# Datamodel for the flow network

from collections import OrderedDict
from numpy import inf, nan

class NetworkElement(object):
    def __init__(self):
        pass

# Node types ...
class Node(NetworkElement):
    """
    Generic network node super class.
    """
    def __init__(self, id):
        super(Node, self).__init__()
        self.id = id
        self.capacity = inf
        self.inarcs = []
        self.outarcs = []
        self.supply = 0

class Sink(Node):
    """
    The sinknode for the problem.
    """
    def __init__(self):
        super(Sink, self).__init__("SINK")


class TargetField(Node):
    """
    Represents a single PFS observation on
    sky with a specific RA/Dec and position angle
    and fiber configuration.
    """
    def __init__(self, id, cost):
        super(TargetField, self).__init__(id)
        self.cost = cost
        self.supply = 1.


class ObservingSlot(Node):
    """
    Represents an observing slot or observin block,
    i.e. a specific time at which PFS executes an observation.
    """
    def __init__(self, id, start_date, end_date):
        super(ObservingSlot, self).__init__(id)
        self.start_date = start_date
        self.end_date = end_date
        self.capacity = 1.


class Network(object):
    """
    Network flow master class.
    """
    def __init__(self):
        self.nodes = {}
        self.arcs = {}

    def add_node(self, node):
        self.nodes[node.id] = node

    def add_arc(self, arc):
        self.arcs[arc.id] = arc
        self.nodes[arc.startnode.id].outarcs.append(arc)
        self.nodes[arc.endnode.id].inarcs.append(arc)


class SurveyPlan(Network):
    def __init__(self):
        super(SurveyPlan, self).__init__()
        # These are just handy for quick access to a certain type of
        # nodes or arcs
        self.targetFields = OrderedDict()
        self.observingSlots = OrderedDict()
        self.sinks = OrderedDict()
        # arcs
        self.visibilities = OrderedDict()
        self.overflowArcs = OrderedDict()

    def add_node(self, node):
        super(SurveyPlan, self).add_node(node)
        if type(node) == TargetField:
            self.targetFields[node.id] = node
        elif type(node) == ObservingSlot:
            self.observingSlots[node.id] = node
        elif type(node) == Sink:
            pass
        else:
            raise Exception("Unknown node type.")

    def get_overflowarcs(self):
        return [a for a in self.arcs if a == OverflowArc]

    def add_arc(self, arc):
        super(SurveyPlan, self).add_arc(arc)       
        if type(arc) == Visibility:
            self.visibilities[arc.id] = arc
        elif type(arc) == OverflowArc:
            self.overflowArcs[arc.id] = arc
        else:
            raise Exception("Unknown arc type: {}".format(type(arc)))


# Arc types ...
class Arc(NetworkElement):
    def __init__(self, startnode, endnode):
        super(Arc, self).__init__()
        self.id = "{}={}".format(startnode.id, endnode.id)
        self.startnode = startnode
        self.endnode = endnode
        self.capacity = 1
        self.cost = 0
        self.flow = 0


class Visibility(Arc):
    """
    Represents a possible target field to observing slot pairing.
    Only if there is an arc between a target field and an observing slot
    the target field is observable during the observing slot.
    """
    def __init__(self, startnode, endnode, cost):
        super(Visibility, self).__init__(startnode, endnode)
        self.cost = cost # cost probably some function of airmass
        self.capacity = 1


class OverflowArc(Arc):
    """
    Generic overflow arc. Tens to be assiciated with a high cost
    to force flow accross the network.
    """
    def __init__(self, startnode, endnode, cost):
        super(OverflowArc, self).__init__(startnode, endnode)
        self.cost = cost
        self.capacity = inf

```

```python
def airmass_to_cost(AM):
    """
    Cost funtion for the visibilities = target field to observing slot pairings.
    Can/should be tuned. Here is dicourage only observations at higher
    airmass.
    """
    return (AM-1)*10.
```

```python
# Build the network that represents the problem.
def buildSurveyPlan(targetfield_table, observingslot_table, visibility_table):
    g = SurveyPlan()

    # Add global sink node
    SINK = Sink()
    g.add_node(SINK)

    # In one go add all arcs for the visibilities
    # and nodes for the target fields
    # and nodes for the observing slots
    for v in visibility_table:
        targetfield_name = v["targetfield"]
        observingslot_name = v["observingslot"]
        
        # add node for the target field if this has not happened before
        if not targetfield_name in g.targetFields:
            # look up cost
            targetfield_cost = targetfield_table["cost"][targetfield_table["name"] == targetfield_name][0]
            targetField = TargetField(targetfield_name, targetfield_cost)
            # add node to graph
            g.add_node(targetField)
        else:
            targetField = g.targetFields[targetfield_name]

        # add node for the observing slot if this has not happened before
        if not observingslot_name in g.observingSlots:
            # look up start date and end data
            start_date = observingslot_table["start"][observingslot_table["name"] == observingslot_name][0]
            end_date = observingslot_table["end"][observingslot_table["name"] == observingslot_name][0]
            observingSlot = ObservingSlot(observingslot_name, start_date, end_date)
            # add node to graph
            g.add_node(observingSlot)
        else:
            observingSlot = g.observingSlots[observingslot_name]

        # finally add the visibility arc between target field and observing slot 
        # and assign cost 
        cost = airmass_to_cost(v["airmass"])
        visibility = Visibility(targetField, observingSlot, cost)
        g.add_arc(visibility)

    # add overflow arcs from target fields to SINK
    # and set cost
    for targetfield_name in g.targetFields:
        targetField = g.targetFields[targetfield_name]
        overflowArc = OverflowArc(targetField, SINK, targetField.cost)
        g.add_arc(overflowArc)
        
    return g

g =  buildSurveyPlan(targetfield_table, observingslot_table, visibility_table)
```

```python
# this is a routine to plot the corresponding
# networkgraph for the problem.
from matplotlib import pyplot as plt
% matplotlib inline
import matplotlib.lines as mlines
from matplotlib import patches as mpatches
from matplotlib import collections


def plotSurveyPlan(g, LABELFLOWS=False, ALTLABELS=False):

    def nlabel(ax, xy, text):
        y = xy[1] - 0.  # shift y-value for label so that it's below the artist
        x = xy[0] + 0.015  # shift x-value for label so that it's next to the artist
        ax.text(x, y, text, family='sans-serif', ha='left', va='center', fontsize=6)

    def plotNodes(ax, nodes, x, color, label, allnodes):
        # targets
        N = len(nodes)
        plt.text(x, 0., label, ha='center', va='top')
        if N == 0:
            return
        grid = [[x, 1. - (float(i)+1.)/float(N+1)] for i in range(N)]
        patches = []
        for i, (id, n) in enumerate(nodes.items()):
            n.px, n.py = grid[i]
            
            circle = mpatches.Circle(grid[i], .01, facecolor=color, edgecolor='grey', lw=1.)
            patches.append(circle)
            nlabel(ax, grid[i], id)

        collection = collections.PatchCollection(patches, match_original=True)

        ax.add_collection(collection)
        allnodes += list(nodes.keys())
        

    def plotArcs(ax, g, allnodes, LABELFLOWS, ALTLABELS):
        for a in g.arcs.values():
            n1, n2 = a.startnode, a.endnode
            if n1.id in allnodes and n2.id in allnodes:
                # add a line
                x, y = np.array([[n1.px, n2.px], [n1.py, n2.py]])
                lw = a.flow * 2 + 1.
                c = 'k'
                rflw = a.flow
                if rflw < 1e-6:
                    rflw = 0.
                if rflw > 0.:
                    c = 'blue'
                line = mlines.Line2D(x, y, lw=lw, alpha=.3, zorder=0, c=c)
                if LABELFLOWS:
                    l = a.id
                    if ALTLABELS is not None:
                        l = ALTLABELS[a.id]
                    ax.text((n1.px+n2.px)/2., (n1.py+n2.py)/2., "{}:{}".format(l, rflw), size=8)
                ax.add_line(line)


    fig = plt.figure(figsize=[15, 15])
    # science targets
    ax = plt.subplot(111)  # note we must use plt.subplots, not plt.subplot
    allnodes = []  # keep track of nodes actually contained in the plot
    plotNodes(ax, nodes=g.targetFields, x=0.2, label="target fields", color="#19967d", allnodes=allnodes)
    plotNodes(ax, nodes=g.observingSlots, x=0.6, label="observing slots", color="#99d5ca", allnodes=allnodes)
    plotArcs(ax, g, allnodes, LABELFLOWS, ALTLABELS)
    ax.set_ylim([-.0, 1.0])
    ax.set_xlim([-.0, 1.0])
    plt.axis('off')
    plt.savefig("observingplan.pdf")
```

```python
from __future__ import print_function
from __future__ import absolute_import
import pulp
from collections import namedtuple
from collections import OrderedDict
import numpy as np
import time

def buildLPProblem(g, name="MinCostFlowTest"):
    """
    Build the LP problem using guropipy's native interface.
    """
    from gurobipy import Model, quicksum, GRB
    
    start_time = time.time()

    m = Model(name)

    flows = {}

    def addFlow(m, flows, n, l=0, u=GRB.INFINITY, vtype=GRB.INTEGER):
        f = m.addVar(vtype=vtype, name=n, lb=l, ub=u)
        flows[n] = f
              
    for arc in g.arcs.values():
        if arc.endnode.id == "SINK":
            addFlow(m, flows, arc.id, 0, 1e12)
        else:
            addFlow(m, flows, arc.id, vtype=GRB.BINARY)  # capacity of one

    # Now add constraints: 
    #  At every targetField node, supply = sum(outflows)
    for tnfid, tfn in g.targetFields.items():
        S = tfn.supply
        m.addConstr( quicksum([flows[a.id] for a in tfn.outarcs]) == S )
        
    # At every observing slot, sum(inflow) = capacity (=1)
    for osfid, osn in g.observingSlots.items():
        C = osn.capacity
        m.addConstr( quicksum([flows[a.id] for a in osn.inarcs]) <= C )


    print("Building cost equation ...")
    # attribute cost to the flows along the overflow arcs
    # Cost occurs in two ways for now, either by
    # flow occuring from a targetClass to the sink node (targets do not get observed at all)
    # or flow from a trget node to the sink node (target was only partially observed)
    cost = m.addVar(vtype=GRB.CONTINUOUS, name="cost")

    m.addConstr( cost == quicksum([a.cost * flows[a.id] for a in g.overflowArcs.values()])\
        + quicksum([a.cost * flows[a.id] for a in g.visibilities.values()]) \
               )
    
    # This sets the cost as objective function for the optimisation.
    m.setObjective(cost, GRB.MINIMIZE)
    time_to_finish = time.time() - start_time
    print(" Time to completion: {:.2f} s".format(time_to_finish))
    
    return m, flows, cost

def setflows(m, g, flows):
    """
    Take the solution of the LP model and sets all the flows in the survey plan 
    graph according to that solution.
    """
    flows_sol = m.getAttr('X', flows)
    for a in g.arcs.values():
        k = '{}={}'.format(a.startnode.id, a.endnode.id)
        if k in flows_sol:
            a.flow = flows_sol[k]
```

```python
m, flows, cost = buildLPProblem(g)
```

```python
import random

summary = ""
t = 4
print("##########################################")
print("Solving with {} threads ...".format(t))
print("##########################################")

m.Params.timelimit = 600.0
m.Params.Threads = t
m.Params.MIPGap = 1e-4

# solve it!
random.seed(42)

# Solve problem!
summary += pp("Solving LP problem ...")
start_time = time.time()

m.optimize()

time_to_solve = time.time() - start_time
summary += pp("Solve status is [{}].".format( m.status ))
summary += pp("Time to solve: {:.4e} s".format(time_to_solve))

s = "{} {}\n".format(m.Params.Threads, time_to_solve)
```

```python
# take the optimal solution flows
# and sets the corresponding arc flow properties back
# in the network graph. This is mostly
# for the ease of inspection of the solution
# and for plotting.
setflows(m, g, flows)
```

```python
# plot the survey plat and the solution
plotSurveyPlan(g)
```

```python
# Finally produce a table for the solution.
# find which target fields got assignd to which observing slots

solution = Table(names=["observingslot","targetfield"], dtype=["U14","U14"])
for s in g.observingSlots.values():
    observingslot_name = s.id 
    targetfield_name = ""
    for a in s.inarcs:
        if a.flow  == 1.:
            targetfield_name = a.startnode.id
    solution.add_row ([observingslot_name, targetfield_name])
```
